```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json

# Import API key
from config import OWM_API_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
#output_data_file = "cities.csv"
#cities_df=pd.read_csv(output_data_file)

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)

```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
city_names=[]
latitudes=[]
longitudes=[]
cloudiness=[]
dates=[]
max_temps=[]
humidities=[]
wind_speeds=[]
countries=[]
```

```python
from time import sleep

responses=[]
# define the base url
url = "http://api.openweathermap.org/data/2.5/weather?"
units="imperial"

for city in cities:
    time.sleep(1)
    try:
        print(f"Processing record {cities.index(city)}| {city}")
    
        query_url = url + "appid=" + OWM_API_key + "&q=" + city + "&units=" + units
    
        weather_data=requests.get(query_url).json()
        responses.append(weather_data)
    
        country=weather_data["sys"]["country"]
        countries.append(country)
    
        wind_speed=weather_data["wind"]["speed"]
        wind_speeds.append(wind_speed)
    
        max_temp=weather_data["main"]["temp_max"]
        max_temps.append(max_temp)
    
        humidity=weather_data["main"]["humidity"]
        humidities.append(humidity)
    
        cloudy=weather_data["clouds"]["all"]
        cloudiness.append(cloudy)
    
        date=weather_data["dt"]
        dates.append(date)
    
        latitude=weather_data["coord"]["lat"]
        latitudes.append(latitude)
    
        longitude=weather_data["coord"]["lon"]
        longitudes.append(longitude)
    
        city_names.append(city)
    except:
        print("City not found. Skipping....")
    
```

```python
#check to see if all lists have the same number of values and that there are more than 500 entries
print(len(city_names))
print(len(latitudes))
print(len(longitudes))
print(len(cloudiness))
print(len(dates))
print(len(max_temps))
print(len(humidities))
print(len(wind_speeds))
print(len(countries))
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.

```python
categories={"City":city_names,"Cloudiness":cloudiness,"Country":countries,"Date":dates,"Humidity":humidities,
           "Lat":latitudes,"Lng":longitudes,"Max Temp":max_temps,"Wind Speed":wind_speeds}
summary_cities=pd.DataFrame(categories)

summary_cities.to_csv('summary_cities.csv',index=False)
summary_cities.head()
```

## Linear Regression

```python
a=0
b=0
def LinearRegression(x,y):
    # Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
    #Calculate the linear regression values to be plotted against weight
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
    regress_values = x * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    #plot the regression values to form the regression line
    plt.plot(x,regress_values,"r-")
    plt.annotate(line_eq,(a,b),fontsize=15,color="red");
    
    print(f"The r-squared is: {rvalue}")
    print(line_eq)
```
